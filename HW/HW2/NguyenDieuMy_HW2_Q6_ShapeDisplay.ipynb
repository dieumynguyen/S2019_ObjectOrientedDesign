### I. Abstract class Shape and its derived subclasses

```java
/* Abstract superclass Shape */

public abstract class Shape implements Comparable<Shape>{
    // Fields aka attributes common to all shapes 
    String name;           // Type of shape
    int x;                 // Starting x-position
    int y;                 // Starting y-position
    private int z_order;   // Order for displaying
    
    // Constructor 
    public Shape(String name, int z_order, int x, int y) {
        this.name = name;
        this.z_order = z_order;  
        this.x = x;
        this.y = y;
    }
    
    // Methods common to all shapes
    public int getZ() {
        return z_order;
    }
    
    public void display() {
        System.out.println(String.format("Displaying a %s...", name));
        System.out.println(String.format("\t z-order: %s", z_order));
        System.out.println(String.format("\t xy: (%s, %s)", x, y));
    }
    
    public String toString() {
        return (String.format("(%s, %s, %s, %s)", name, z_order, x, y));
    }
    
    @Override
    public int compareTo(Shape shape) {
        return z_order - shape.getZ();
    }
}
```

```java
/* Concrete class inheriting from Shape #1: Circle */

public class Circle extends Shape {
    // Fields custom for this concrete class 
    int radius;
    
    // Constructor
    public Circle(String name, int z_order, int x, int y, int radius) {
        super(name, z_order, x, y);
        this.radius = radius;
    }
    
    // Methods
    public void display() {
        super.display();
        System.out.println(String.format("\t radius: %s", radius));
    }
    
    public String toString() {
        return (String.format("(%s, %s, %s, %s, %s)", name, getZ(), x, y, radius));
    }
}
```

```java
/* Concrete class inheriting from Shape #2: Square */

public class Square extends Shape {
    // Fields custom for this concrete class 
    int width;
    int height;
    
    // Constructor
    public Square(String name, int z_order, int x, int y, int width, int height) {
        super(name, z_order, x, y);
        this.width = width;
        this.height = height;
    }
    
    // Methods
    public void display() {
        super.display();
        System.out.println(String.format("\t width: %s", width));
        System.out.println(String.format("\t height: %s", height));
    }    
    
    public String toString() {
        return (String.format("(%s, %s, %s, %s, %s, %s)", name, getZ(), x, y, width, height));
    }
}
```

```java
/* Concrete class inheriting from Shape #3: Triangle */

public class Triangle extends Shape {
    // Fields custom for this concrete class 
    int base;
    int height;
    
    // Constructor
    public Triangle(String name, int z_order, int x, int y, int base, int height) {
        super(name, z_order, x, y);
        this.base = base;
        this.height = height;
    }
    
    // Methods
    public void display() {
        super.display();
        System.out.println(String.format("\t base: %s", base));
        System.out.println(String.format("\t height: %s", height));
    }
        
    public String toString() {
        return (String.format("(%s, %s, %s, %s, %s, %s)", name, getZ(), x, y, base, height));
    }
}
```

### II. A Collection class to contain all shapes, sort, and display them

```java
public class Collection {    
    // Methods
    public void sort(Shape[] shapes) {
        // Sort by z_order, ascending order
        Arrays.sort(shapes);
    }
    
    public void display(Shape[] shapes) {
        // Display all shapes by z_order
        for (Shape s : shapes) {
            s.display();
        }
    }        
}
```

### III. Main program

```java
public class makeDatabase {
    // Fields
    int num_type;
    int num_shapes;

    // Constructor
    public makeDatabase(int num_type, int num_shapes) {
        this.num_type = num_type;
        this.num_shapes = num_shapes;  
    }
    
    // Methods
    public List<Integer> createRandomZ() {
        List<Integer> random_zs = new ArrayList<>();
        for (int j=0; j<num_shapes; j++) random_zs.add(j);
        Collections.shuffle(random_zs);
        return random_zs;
    }
    
    public Shape[] createShapesArray(List<Integer> random_zs) {
        Shape[] shapes = new Shape[num_shapes];
        for (int i=0; i<num_shapes; i++) {    
            // Fields for all shapes
            int randomZ = random_zs.get(i);
            int randomX = random.nextInt(10);
            int randomY = random.nextInt(10);
            int randomRadius = random.nextInt(100);
            int randomBase = random.nextInt(50);
            int randomWidth = random.nextInt(50);
            int randomHeight = random.nextInt(50);

            // Instantiate objects of random derived Shape classes
            int randomChoice = random.nextInt(num_types);
            if (randomChoice == 0) {
                shapes[i] = new Circle("circle", randomZ, randomX, randomY, randomRadius);
            } else if (randomChoice == 1) {
                shapes[i] = new Square("square", randomZ, randomX, randomY, randomWidth, randomHeight);
            } else {
                shapes[i] = new Triangle("triangle", randomZ, randomX, randomY, randomBase, randomHeight);
            }
        }
        return shapes;
    }
}
```

```java
public static void main(String... args) {
    Random random = new Random();

    /********** Create a shape database **********/
    int num_types = 3;
    int num_shapes = 10;
    
    makeDatabase database = new makeDatabase(num_types, num_shapes);
    random_zs = database.createRandomZ();
    shapes = database.createShapesArray(random_zs);

    /********** Create a Collection object, sort, display all shapes **********/
    System.out.println("-------- SHAPE DISPLAY -------- \n");
    System.out.println(String.format("Number of shapes in database: %s \n", num_shapes));

    Collection coll = new Collection();
    coll.sort(shapes);
    coll.display(shapes);
}

main();
```

```java
/* Abstract class Employee with 1 constructor, 3 fields, 7 methods */

public abstract class Employee {
    /* Fields aka attributes */
    private String name;
    private String address;
    private int number;
    
    /* Constructor */
    public Employee(String name, String address, int number) {
        System.out.println("\n Constructing an Employee");
        this.name = name;
        this.address = address;
        this.number = number;
    }
    
    /* Methods */
    public double computePay() {
        System.out.println("Inside Employee computePay");
        return 0.0;
    }
    
    public void mailCheck() {
        System.out.println("Mailing a check to " + this.name + " " + this.address);
    }
    
    public String toString() {
        return name + " " + address + " " + number;
    }
    
    public String getName() {
        return name;
    }
    
    public String getAddress() {
        return address;
    }
    
    public void setAddress(String newAddress) {
        address = newAddress;
    }
    
    public int getNumber() {
        return number;
    }
}
```

```java
/* Instantiate the Employee abstract class: 
Expect error because abstract classes cannot be instantiated! */ 

public class AbstractDemo {
    public static void main(String[] args) {
        /* The following is not allowed */
        Employee e =  new Employee("Hello Kitty", "Canada", 100);
        System.out.println("\n Call mailCheck for Employee e...");
        e.mailCheck();
    }
}
```

```java
/* Inherit the abstract Employee class */

public class Salary extends Employee {
    /* Fields custom for this concrete class */
    private double salary;
    
    /* Constructor */
    public Salary(String name, String address, int number, double salary) {
        /* Include inherited fields from superclass */    
        super(name, address, number);
        
        setSalary(salary);
    }
    
    /* Overriding mailCheck of superclass? */
    public void mailCheck() {
        System.out.println("Within mailCheck of Salary class");
        System.out.println("Mailing check to " + getName() + " with salary " + salary);        
    }
    
    public double getSalary() {
        return salary;
    }
    
    public void setSalary(double newSalary) {
        if (newSalary >= 0.0) {
            salary = newSalary;
        }
    }
    
    public double computePay() {
        System.out.println("Computing salary pay for " + getName());
        return salary / 52;
    }
}
```

```java
/* Cannot instantiate Employee class, but can instantiate Salary class
Salary instances can access all 3 fields and 7 methods of Employee */

public class AbstractDemo {
    public static void main(String ... args) {
        Salary s = new Salary("Penn", "Seattle, WA", 1, 5000.0);
        System.out.println("Call mailCheck using Salary reference --");
        s.mailCheck();
        
        Employee e = new Salary("Teller", "Boston, MA", 2, 7000.0);
        System.out.println("Call mailCheck using Employee reference --");
        e.mailCheck();
    }
}

AbstractDemo test = new AbstractDemo();
// String[] args = {};
// test.main(args);
test.main();
```
